```python
customer_master = pd.read_csv('/content/testData/No1/customer_master.csv')
item_master = pd.read_csv('/content/testData/No1/item_master.csv')
print(customer_master.head())
print(item_master.head())
```

```python
customer_item_data = pd.merge(join_data, customer_master, on="customer_id", how="left")
customer_item_data = pd.merge(join_data, item_master, on="item_id", how="left")
customer_item_data.head()
```

```python
customer_item_data['price'] = customer_item_data['quantity'] * customer_item_data['item_price']
customer_item_data[["quantity", "item_price", "price"]].head()
```

```python
# けんさんデータ
customer_item_data["price"].sum() == transaction["price"].sum()
```

```python
# 各種統計量を把握
customer_item_data.isnull().sum()
customer_item_data.describe()
```

```python
customer_item_data.dtypes
```

```python
customer_item_data["payment_date"] = pd.to_datetime(customer_item_data["payment_date"])
customer_item_data["payment_month"] = customer_item_data["payment_date"].dt.strftime("%Y%m")
# 集計しやすく変換↓
customer_item_data["payment_month"] = customer_item_data["payment_date"].dt.to_period("M")
customer_item_data[["payment_date", "payment_month"]].head()
```

```python
## 月毎の算出
# customer_item_data.groupby("payment_month").sum()["price"]

# price列を念のため数値型に変換
customer_item_data["price"] = pd.to_numeric(customer_item_data["price"], errors="coerce")

# 月ごとの合計
monthly_sum = customer_item_data.groupby("payment_month")["price"].sum()
print(monthly_sum)
```

```python
customer_item_data.groupby(["payment_month", "item_name"]).sum()["price", "quantity"]
```

```python
pd.pivot_table(customer_item_data, index='item_name', columns='payment_month', values=['price', 'quantity'], aggfunc='sum')
```

```python
graph_data = pd.pivot_table(customer_item_data, index='payment_month', columns='item_name', values='price', aggfunc='sum')
graph_data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(list(graph_data.index.to_timestamp()), graph_data["PC-A"], label="PC-A")
plt.plot(list(graph_data.index.to_timestamp()), graph_data["PC-B"], label="PC-B")
plt.plot(list(graph_data.index.to_timestamp()), graph_data["PC-C"], label="PC-C")
plt.plot(list(graph_data.index.to_timestamp()), graph_data["PC-D"], label="PC-D")
plt.plot(list(graph_data.index.to_timestamp()), graph_data["PC-E"], label="PC-E")
plt.legend()
```
